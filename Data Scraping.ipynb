```python
from bs4 import BeautifulSoup as bs4
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_Indonesia'
page = requests.get(url)
soup = bs4(page.text, 'html')

title = soup.find('h1', class_ = 'firstHeading mw-first-heading')
title = title.text

table = soup.find_all('table')[0]
table_head = table.find_all('th')
print(table_head)
```

HEADERS

```python
headers = []
for header in table_head:
    headers2 = header.text.strip()
    headers.append(headers2)
print(headers)

# OR
# table_heads = [header.text.strip() for header in table_head]
# print(table_heads)

DF = pd.DataFrame(columns= headers)
DF
```

DATAS

```python
rows = table.find_all('tr')
for row in rows[1:]:
    table_data = row.find_all('td')
    datas = [data.text.strip() for data in table_data]

    lenght = len(DF)
    DF.loc[lenght]= datas

DF
```

```python
DF.to_csv(r'YOUR PATH/List of largest companies in Indonesia.csv', index = False)
# Ganti YOUR PATH dengan path folder yang ingin disimpan
```
